```python
import pandas as pd
```

```python
df = pd.read_csv('/content/airpar.csv')
```

```python
df
```

```python
df_error = df[(df['time'].isnull())]
df_error['time'] = df_error['Day']
df_error['Day'] = '0일'

null_time = df[(df['time'].isnull())]
df = df.drop(index=null_time.index).reset_index(drop=True)

df_com = pd.concat([df, df_error], axis=0)
df_com = df_com.reset_index(drop=True)
```

```python
df_com1 = df_com[['Date', 'Airline', 'Departure', 'Arrival', 'Day', 'Payment', 'Price']]
df_com2 = df_com.time.str.split(',', expand = True)
df_com2.columns = ['Through', 'Time']
```

```python
df_co = pd.concat([df_com1, df_com2], axis = 1)
```

```python
df_coerpr = df_co[(df_co['Price'] == '할인')]
df_co = df_co.drop(index=df_coerpr.index).reset_index(drop=True)
```

```python
df_co
```

```python
def extract_price(price_str):
    price_str = price_str.replace('편도 ', '')
    price_str = price_str.replace(',', '')
    price_str = price_str.replace('원~', '')
    price = int(price_str)
    return price
```

```python
df_co['Price'] = df_co['Price'].apply(extract_price)

df_co['Departure_Time'] = df_co['Departure'].str.extract(r'(\d{2}:\d{2})[^:]+$')
df_co['Departure'] = df_co['Departure'].str.replace('[^a-zA-Z\s]', '')

df_co['Arrival_Time'] = df_co['Arrival'].str.extract(r'(\d{2}:\d{2})[^:]+$')
df_co['Arrival'] = df_co['Arrival'].str.replace('[^a-zA-Z\s]', '')
```

```python
df_co
```

```python
df_clean = df_co.copy()
```

```python
df_clean
```
