```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/airkor.csv')
df2 = pd.read_csv('/content/airkor_pn.csv')
df3 = pd.read_csv('/content/airkor_BC.csv')
df4 = pd.read_csv('/content/airkor_FC.csv')
df5 = pd.read_csv('/content/airpar.csv')
df6 = pd.read_csv('/content/airpar_pn.csv')
df7 = pd.read_csv('/content/airpar_BC.csv')
df8 = pd.read_csv('/content/airpar_FC.csv')
```

```python
def feature_engineering(df):

  df_error = df[(df['time'].isnull())]
  df_error['time'] = df_error['Day']
  df_error['Day'] = '0일'

  null_time = df[(df['time'].isnull())]
  df = df.drop(index=null_time.index).reset_index(drop=True)

  df = pd.concat([df, df_error], axis=0)
  df = df.reset_index(drop=True)
  
  null_price = df[(df['Price'] == '할인')]
  df = df.drop(index=null_price.index).reset_index(drop=True)

  return df
```

```python
df_clean1 = feature_engineering(df1)
df_clean2 = feature_engineering(df2)
df_clean3 = feature_engineering(df3)
df_clean4 = feature_engineering(df4)
df_clean5 = feature_engineering(df5)
df_clean6 = feature_engineering(df6)
df_clean7 = feature_engineering(df7)
df_clean8 = feature_engineering(df8)
```

```python
def feature_engineering1(df):

  df['Departure_Time'] = df['Departure'].str.extract(r'(\d{2}:\d{2})[^:]+$')
  df['Departure'] = df['Departure'].str.replace('[^a-zA-Z\s]', '')

  df['Arrival_Time'] = df['Arrival'].str.extract(r'(\d{2}:\d{2})[^:]+$')
  df['Arrival'] = df['Arrival'].str.replace('[^a-zA-Z\s]', '')

  return df
```

```python
df_clean1 = feature_engineering1(df_clean1)
df_clean2 = feature_engineering1(df_clean2)
df_clean3 = feature_engineering1(df_clean3)
df_clean4 = feature_engineering1(df_clean4)
df_clean5 = feature_engineering1(df_clean5)
df_clean6 = feature_engineering1(df_clean6)
df_clean7 = feature_engineering1(df_clean7)
df_clean8 = feature_engineering1(df_clean8)
```

```python
def extract_price(price_str):
    price_str = price_str.replace('편도 ', '')
    price_str = price_str.replace(',', '')
    price_str = price_str.replace('원~', '')
    price = int(price_str)
    return price
```

```python
df_clean1['Price'] = df_clean1['Price'].apply(extract_price)
df_clean2['Price'] = df_clean2['Price'].apply(extract_price)
df_clean3['Price'] = df_clean3['Price'].apply(extract_price)
df_clean4['Price'] = df_clean4['Price'].apply(extract_price)
df_clean5['Price'] = df_clean5['Price'].apply(extract_price)
df_clean6['Price'] = df_clean6['Price'].apply(extract_price)
df_clean7['Price'] = df_clean7['Price'].apply(extract_price)
df_clean8['Price'] = df_clean8['Price'].apply(extract_price)
```

```python
def feature_engineering2(df):

  df_com1 = df[['Date', 'Airline', 'Departure', 'Arrival', 'Day', 'Payment', 'Price', 'Departure_Time', 'Arrival_Time']]
  df_com2 = df.time.str.split(',', expand = True)
  df_com2.columns = ['Through', 'Time']
  df = pd.concat([df_com1, df_com2], axis = 1)

  return df
```

```python
df_clean1 = feature_engineering2(df_clean1)
df_clean1['Seat'] = '일반'

df_clean2 = feature_engineering2(df_clean2)
df_clean2['Seat'] = '프리미엄일반'

df_clean3 = feature_engineering2(df_clean3)
df_clean3['Seat'] = '비즈니스'

df_clean4 = feature_engineering2(df_clean4)
df_clean4['Seat'] = '퍼스트클래스'

df_clean5 = feature_engineering2(df_clean5)
df_clean5['Seat'] = '일반'

df_clean6 = feature_engineering2(df_clean6)
df_clean6['Seat'] = '프리미엄일반'

df_clean7 = feature_engineering2(df_clean7)
df_clean7['Seat'] = '비즈니스'

df_clean8 = feature_engineering2(df_clean8)
df_clean8['Seat'] = '퍼스트클래스'
```

```python
df_list_pariskorea = [df_clean1, df_clean2, df_clean3, df_clean4]
df_list_koreaparis = [df_clean5, df_clean6, df_clean7, df_clean8]
```

```python
df_pariskorea = pd.concat(df_list_pariskorea, ignore_index=True)
df_koreaparis = pd.concat(df_list_koreaparis, ignore_index=True)
```
