```python
import psycopg2
import csv
```

```python
host = '34.64.57.43'
user = 'khs03'
password = '1234!'
database = 'ai-17-seven'

connection = psycopg2.connect(
    host=host,
    user=user,
    password=password,
    database=database)

cur = connection.cursor()

cur.execute("DROP TABLE IF EXISTS attraction;") 

cur.execute("""
CREATE TABLE attraction(
  Id SERIAL PRIMARY KEY,
  name VARCHAR(200),
  category VARCHAR(200),
  score FLOAT,
  priceif VARCHAR(200),
  time VARCHAR(200),
  price INTEGER,
  priceinfomation VARCHAR(200)
)
""")

with open('C:/Users/admin/Desktop/attraction_clean_price.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        try:
            cur.execute("""
                INSERT INTO attraction (
                  name, category, score, priceif,
                  time, price, priceinfomation) 
                VALUES (%s, %s, %s, %s, %s, %s, %s)
            """, row)
        except IndexError:
            print(f"Skipping row {row}")
            continue

connection.commit()
cur.close()
connection.close()
```
