```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/hotel_data_6.csv')
df2 = pd.read_csv('/content/hotel_data_7.csv')
df3 = pd.read_csv('/content/hotel_data_8.csv')

df_list = [df1, df2, df3]
df = pd.concat(df_list, ignore_index = True)
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(['page','date_time_now'], axis = 1 )
df = df.drop_duplicates()
df = df.drop(['locations','check_out'], axis = 1 )
```

```python
df.info()
```

```python
df_error = df[(df['review_counts'] <= 0) | (df['prices'].isnull()) | (df['discount_prices'].isnull())] 
df = df.drop(index=df_error.index).reset_index(drop=True)
```
