```python
import pandas as pd
```

```python
df = pd.read_csv('/content/attraction_clean.csv')
```

```python
df
```

```python
df.info()
```

```python
df['price'] = df['price'].astype(str)
```

```python
df['price_numeric'] = df['price'].apply(lambda x: int(x.split(' ')[0].replace(',', '')))
df['price_text'] = df['price'].apply(lambda x: ' '.join(x.split(' ')[1:]))
```

```python
df.to_csv('attraction_clean_price.csv', index=False)
```
